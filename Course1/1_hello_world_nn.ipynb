# Introduction to TensorFlow for Artificial Intelligence, Machine Learning, and Deep Learning

## 1주차 강의
The ‘Hello World’ of neural networks

- Dense:A layer of connected neurons 
- Loss function:Measures how good the current ‘guess’ is 
- Optimizer:Generates a new and improved guess
- Convergence: The process of getting very close to the correct answer  
- Sequential: successive layers are defined in sequence
-  fit: trains the neural network to fit one set of values to another

## Imports
importing TensorFlow and calling it `tf` for convention and ease of use.

import a library called `numpy` which helps to represent data as arrays easily and to optimize numerical operations.

The framework you will use to build a neural network as a sequence of layers is called `keras` so you will import that too.

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

## Define and Compile the Neural Network

neuron 1개, input shape이 1인 neural network 생성

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])]) # Build a simple Sequential model
model.compile(optimizer='sgd', loss='mean_squared_error') # Compile the model 
```

## Providing the Data

```python
# Declare model inputs and outputs for training
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

## Training the Neural Network

```python
model.fit(xs, ys, epochs=500) # 모델 학습
```

```python
print(model.predict([10.0])) # 예측
```

주어진 데이터가 Y = 2X - 1를 따르기 때문에 model.predit()가 19를 반환할 것이라고 생각할 수도 있다. 하지만, 실제로 19에 매우 가까운 값을 반환하지만 정확히 19는 아니다.<br>

**왜?**
- 매우 적은 데이터를 사용하여 훈련했다. 
주어진 데이터는 6개인데, 이 6개 점은 선형이지만 모든 X에 대해 관계가 Y = 2X - 1이라는 보장은 없다. 
- X가 10인 경우 Y가 19일 가능성이 매우 높지만 신경망은 양수가 아니다. 따라서 Y에 대한 현실적인 값을 알아낼 것이다.
- 신경망을 사용할 때 모든 것에 대한 답을 찾으려고 할 때 확률을 다룬다. 그래서 적합하도록 답변을 처리하는 방법을 조정해야 한다.
